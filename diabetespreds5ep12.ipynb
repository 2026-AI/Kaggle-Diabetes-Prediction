```python
# x=df_data.drop('diagnosed_diabetes',inplace=False)
# y=df_data['diagnosed_diabetes']
# diagnosed_diabetes
x = df_data_train.drop(columns=['diagnosed_diabetes'])
y = df_data_train['diagnosed_diabetes']
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
```

```python
# one hot encoding for categorical columns
categorical_cols = [
    'gender',
    'ethnicity',
    'education_level',
    'income_level',
    'smoking_status',
    'employment_status'
]
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42, stratify=y
)
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(handle_unknown='ignore', drop='first'),
         categorical_cols)
    ],
    remainder='passthrough'
)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
```

```python
model = Pipeline([
    ('prep', preprocessor),
    ('clf', LogisticRegression(
        solver='saga',
        max_iter=5000,
        n_jobs=-1
    ))
])
```

```python
model.fit(X_train, y_train)
```

```python
val_predictions = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report

print("Accuracy:", accuracy_score(y_test, val_predictions))
print("Precision:", precision_score(y_test, val_predictions))
print("Recall:", recall_score(y_test, val_predictions))
print("F1 Score:", f1_score(y_test, val_predictions))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, val_predictions))

print("\nClassification Report:")
print(classification_report(y_test, val_predictions))
```

```python
# predicting on actual test dataset
test_predictions = model.predict(test)
```

```python
submission = pd.DataFrame({
    'id': sub['id'],
    'diagnosed_diabetes': test_predictions
})
```

```python
submission.to_csv("submission.csv", index=False)
```
